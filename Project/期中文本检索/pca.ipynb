```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
```

```python
newsvec = pd.read_csv('newsvec.csv')
newsvec = newsvec.drop(columns='Unnamed: 0')
newsvec
```

```python
# 提取主成分，降维
pca = PCA(n_components=50)
news_pca = pca.fit_transform(newsvec)
```

```python
pcadf = pd.DataFrame(news_pca)
pcadf.to_csv("vecpca2.csv", sep=',')
```

```python
# 计算余弦相似度
def similar(news_pca):
    ret = np.zeros((newsvec.shape[0], newsvec.shape[0]))
    for i in range(newsvec.shape[0]):
        for j in range(newsvec.shape[0]):
            x = news_pca[i]
            y = news_pca[j]
            cosine = np.dot(x,y)/(np.linalg.norm(x)*np.linalg.norm(y))
            ret[i][j] = cosine
    return ret

cos_sim = pd.DataFrame(similar(news_pca), index=range(2225), columns=range(2225))
cos_sim
```

```python
# 保存
cos_sim.to_csv("cos_sim.csv")
```
